# **Exploratory Data Analysis**

```python
def styling_table_contents(table_input):
    # Define the CSS styles for center alignment
    styles = [
        {'selector': 'th', 'props': [('text-align', 'center')]},  # For column names
        {'selector': 'td', 'props': [('text-align', 'center')]}   # For table content
    ]

    # Apply the styles to the DataFrame
    styled_table = table_input.style.set_table_styles(styles).set_properties(**{'text-align': 'center'})

    # Return the styled table
    return styled_table
```

1. Keterkaitan Kawasan Hutan dan Pemenuhan Bahan Baku

a. KAWASAN HUTAN

- Area Kawasan Hutan Paling Luas berdasarkan provinsi

```python
# Group by provinsi and calculate the sum of the area
area_grouped_by_provinsi = data_kawasan_hutan.groupby('provinsi')['area'].sum().reset_index()

# Set the desired formatting for the values
pd.options.display.float_format = '{:,.2f}'.format

# Sort the values in descending order
area_grouped_by_provinsi_sorted = area_grouped_by_provinsi.sort_values(by='area', ascending=False)

# Rename the columns
area_grouped_by_provinsi_sorted = area_grouped_by_provinsi_sorted.rename(columns={'provinsi': 'Provinsi', 'area': 'Luas Tanah (Ha)'})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(area_grouped_by_provinsi_sorted)

# Format the 'Luas Tanah' column
formatted_table = styled_data.format({'Luas Tanah (Ha)': '{:,.2f}'})
formatted_table
```

```python
fig = px.bar(area_grouped_by_provinsi_sorted, x='Provinsi', y='Luas Tanah (Ha)', title='Luas Tanah (Ha) Berdasarkan Provinsi')
fig.update_layout(xaxis_title='Provinsi', yaxis_title='Luas Tanah (Ha)')
fig.show()
```

- Area Kawasan Hutan Paling Luas berdasarkan jenis

```python
# Group by jenis and calculate the sum of the area
area_grouped_by_jenis = data_kawasan_hutan.groupby('jenis')['area'].sum().reset_index()

# Sort the values in descending order
area_grouped_by_jenis_sorted = area_grouped_by_jenis.sort_values(by='area', ascending=False)

# Rename the columns
area_grouped_by_jenis_sorted = area_grouped_by_jenis_sorted.rename(columns={'jenis': 'Jenis Hutan', 'area': 'Luas Tanah (Ha)'})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(area_grouped_by_jenis_sorted)

# Format the 'Luas Tanah' column
formatted_table = styled_data.format({'Luas Tanah (Ha)': '{:,.2f}'})
formatted_table
```

```python
fig = px.bar(area_grouped_by_jenis_sorted, x='Jenis Hutan', y='Luas Tanah (Ha)', title='Luas Tanah (Ha) Berdasarkan Jenis Hutan')
fig.update_layout(xaxis_title='Jenis Hutan', yaxis_title='Luas Tanah (Ha)')
fig.show()
```

b. BAHAN BAKU

- Provinsi yang memiliki bahan baku terbanyak

```python
# Group by provinsi and calculate the sum of the value
bahan_baku_grouped_by_provinsi = data_bahan_baku.groupby('provinsi')['value'].sum().reset_index()

# Sort the values in descending order
bahan_baku_grouped_by_provinsi_sorted = bahan_baku_grouped_by_provinsi.sort_values(by='value', ascending=False)

# Rename the columns
bahan_baku_grouped_by_provinsi_sorted = bahan_baku_grouped_by_provinsi_sorted.rename(columns={'provinsi': 'Provinsi', 'value': 'Bahan Baku'})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(bahan_baku_grouped_by_provinsi_sorted)

# Format the 'Bahan Baku' column
formatted_table = styled_data.format({'Bahan Baku': '{:,.2f}'})
formatted_table
```

```python
fig = px.bar(bahan_baku_grouped_by_provinsi_sorted, x='Provinsi', y='Bahan Baku', title='Bahan Baku Berdasarkan Provinsi')
fig.update_layout(xaxis_title='Provinsi', yaxis_title='Bahan Baku')
fig.show()
```

- Jenis Kayu yang memiliki bahan baku terbanyak

```python
# Group by jenis and calculate the sum of the value
bahan_baku_grouped_by_jenis = data_bahan_baku.groupby('jenis')['value'].sum().reset_index()

# Sort the values in descending order
bahan_baku_grouped_by_jenis_sorted = bahan_baku_grouped_by_jenis.sort_values(by='value', ascending=False)

# Rename the columns
bahan_baku_grouped_by_jenis_sorted = bahan_baku_grouped_by_jenis_sorted.rename(columns={'jenis': 'Jenis Kayu', 'value': 'Bahan Baku'})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(bahan_baku_grouped_by_jenis_sorted)

# Format the 'Bahan Baku' column
formatted_table = styled_data.format({'Bahan Baku': '{:,.2f}'})

formatted_table
```

```python
fig = px.bar(bahan_baku_grouped_by_jenis_sorted, x='Jenis Kayu', y='Bahan Baku', title='Bahan Baku Berdasarkan Jenis Kayu')
fig.update_layout(xaxis_title='Jenis Kayu', yaxis_title='Bahan Baku')
fig.show()
```

c. MERGED DATA

- Merged data kawasan hutan dan bahan baku based on provinsi. Sort by Area

```python
# Merge the data based on the 'provinsi' column
merged_data = pd.merge(area_grouped_by_provinsi, bahan_baku_grouped_by_provinsi, on='provinsi', how='outer')

# Replace NaN values with 0
merged_data['area'] = merged_data['area'].fillna(0)
merged_data['value'] = merged_data['value'].fillna(0)

# Sort the merged dataframe by land area and bahan baku in descending order
merged_data_sorted = merged_data.sort_values(by=['area', 'value'], ascending=False)

# Rename the columns
merged_data_sorted = merged_data_sorted.rename(columns={'provinsi': 'Provinsi', 'area': 'Luas Tanah (Ha)', 'value': 'Bahan Baku'})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(merged_data_sorted)

# Format the 'Bahan Baku' column
formatted_table = styled_data.format({'Luas Tanah (Ha)': '{:,.2f}', 'Bahan Baku': '{:,.2f}'})

formatted_table
```

```python
fig = go.Figure()

# Add bar traces for Luas Tanah (Ha)
fig.add_trace(go.Bar(
    x=merged_data_sorted['Provinsi'],
    y=merged_data_sorted['Luas Tanah (Ha)'],
    name='Luas Tanah (Ha)',
    yaxis='y',
    offset=0,
    width=0.4,
    marker_color='blue'
))

# Add bar traces for Bahan Baku
fig.add_trace(go.Bar(
    x=merged_data_sorted['Provinsi'],
    y=merged_data_sorted['Bahan Baku'],
    name='Bahan Baku',
    yaxis='y2',
    offset=0.4,
    width=0.4,
    marker_color='green'
))

# Set layout
fig.update_layout(
    title='Luas Tanah dan Bahan Baku Berdasarkan Provinsi',
    xaxis=dict(title='Provinsi'),
    yaxis=dict(title='Luas Tanah (Ha)', side='left', showgrid=False, tickformat=',.2f'),
    yaxis2=dict(title='Bahan Baku', side='right', overlaying='y', showgrid=False, tickformat=',.2f'),
    barmode='group',
    legend=dict(x=0.8, y=1),
)

fig.show()
```

- Merged data kawasan hutan dan bahan baku based on provinsi. Membandingkan Bahan Baku lokal dan impor

```python
# Aggregate the "Bahan Baku" for all provinces
total_bahan_baku = merged_data_sorted['Bahan Baku'].sum()
bahan_baku_impor = merged_data_sorted[merged_data_sorted['Provinsi'] == 'Impor Bahan Baku']['Bahan Baku'].values[0]
bahan_baku_lokal = total_bahan_baku - bahan_baku_impor

# Calculate the percentage
percentage_bahan_baku_impor = (bahan_baku_impor / total_bahan_baku) * 100
percentage_bahan_baku_lokal = (bahan_baku_lokal / total_bahan_baku) * 100

# Create a new DataFrame for the comparison
comparison_data = pd.DataFrame({
    'Sumber': ['Bahan Baku Lokal', 'Impor Bahan Baku', 'Total Bahan Baku'],
    'Bahan Baku': [bahan_baku_lokal, bahan_baku_impor, total_bahan_baku],
    'Persentase (%)': [percentage_bahan_baku_lokal, percentage_bahan_baku_impor, 100]
})

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(comparison_data)

# Format the 'Bahan Baku' column
formatted_table = styled_data.format({'Bahan Baku': '{:,.2f}', 'Persentase (%)': '{:,.2f}'})

formatted_table
```

```python
# Filter the comparison data for Impor Bahan Baku and Bahan Baku Lokal
filtered_data = comparison_data[:-1]  # Exclude the last row for Total Bahan Baku

# Create a bar trace for Impor Bahan Baku and Bahan Baku Lokal
bar_trace = go.Bar(
    x=filtered_data['Sumber'],
    y=filtered_data['Bahan Baku'],
    text=filtered_data['Bahan Baku'],
    textposition='auto',
    marker_color=['blue', 'green']
)

# Set layout
layout = go.Layout(
    title='Perbandingan Sumber Pemenuhan Bahan Baku',
    xaxis=dict(title='Sumber Bahan Baku'),
    yaxis=dict(title='Bahan Baku'),
)

# Create a figure and add the trace
fig = go.Figure(data=[bar_trace], layout=layout)

# Show the figure
fig.show()
```

```python
# Merge the data based on the 'provinsi' column
merged_data = pd.merge(data_kawasan_hutan, data_bahan_baku, how='outer')

# Replace NaN values with 0
merged_data['area'] = merged_data['area'].fillna(0)
merged_data['value'] = merged_data['value'].fillna(0)
merged_data['tahun'] = merged_data['tahun'].fillna(0)
merged_data['bulan'] = merged_data['bulan'].fillna(0)
merged_data['jenis'] = merged_data['jenis'].fillna(0)

# Sort the merged dataframe by land area and bahan baku in descending order
merged_data_sorted = merged_data.sort_values(by=['provinsi', 'area'], ascending=False)

# Rename the columns
merged_data_sorted = merged_data_sorted.rename(columns={'provinsi': 'Provinsi', 'tahun': 'Tahun', 'bulan': 'Bulan', 'jenis': 'Jenis','area': 'Luas Tanah (Ha)', 'value': 'Bahan Baku'})

# Convert 'Tahun' column to integer
merged_data_sorted['Tahun'] = merged_data_sorted['Tahun'].astype(int)

# Rearrange the columns
merged_data_sorted = merged_data_sorted[['Tahun', 'Bulan', 'Provinsi', 'Jenis', 'Luas Tanah (Ha)', 'Bahan Baku']]

# Apply styling and formatting to the merged data
styled_data = styling_table_contents(merged_data_sorted)

# Format the 'Bahan Baku' column
formatted_table = styled_data.format({'Luas Tanah (Ha)': '{:,.2f}', 'Bahan Baku': '{:,.2f}'})

formatted_table
```
